## 1. Introduction

The congestion charge in the centre of London was introduced in 2003. People had to pay a few British pounds to enter the area with a vehicle in London during specific hours. The goal of this charge was to reduce the vehicular activities in the centre of London. Therefore, it is interesting to examine if the congestion charge is successful in achieving this goal. This leads to the following research question:

    “How does the congestion charge in London impact the vehicular activities within the congestion charge zone?”

To answer this question, there will be a look into the composition of vehicles within the congestion charge zone and the amount of kilometres travelled by cars within the congestion charge zone. This leads to the following two sub questions:

    1) "To what extent does the congestion charge in London affect the numbers of different vehicle types within the congestion charge zone?"
    2) "To what extent does the congestion charge in London affect the travelled kilometres by cars within the congestion charge zone?"

The time-scale of this research is 2000 till 2021. The congestion charge zone was introduced in 2003, so the research started a few years earlier to look into vehicular activities before the congestion charge zone in London was introduced. After 2003, the charge to drive in the zone has increased every few years. To look into the data till 2021, the effect of the increasing charge could be known.


## 2. Background research and conceptualisation

The congestion charge in London was established at the beginning of 2003, requiring drivers to pay a specific fee to enter a designated area of London during certain weekday hours. Over the years, the cost of the charge has progressively increased (Dunt, 2023), as illustrated in Table 1. The designated zone for this congestion charge encompasses Westminster, Lambeth, Southwark, City of London, Tower Hamlets, Islington and Camden (Tang, 2021).


The primary objective of this congestion charge is to reduce the appeal of various vehicle types, including two wheeled motor vehicles, cars, taxis, buses, coaches, Large Goods Vehicle (LGV) and Heavy Goods Vehicle (HGV) (Beevers & Carslaw, 2005). This reduced appeal is deemed essential in addressing traffic congestion, improving air quality, combating climate change, and reducing noise pollution in London (Leape, 2006).

Numerous studies have demonstrated that the implementation of a congestion charge has the potential to curtail both the number of vehicles and the total distance travelled by vehicles (Börjesson & Kristoffersson, 2015; Eliasson, 2008; Gibson & Carnovale, 2015; May & Milne, 2000; Seik, 2000). As such, the underlying hypothesis posits that with the introduction of the congestion charge in London in 2003, there will be a gradual reduction over time in the number of all vehicle types within the charge zone, as well as a decline in the total kilometres travelled by cars in the charge zone. It is anticipated that the most substantial decline in both vehicle counts, and kilometres travelled will be observed following the charge’s initiation in 2003 and its subsequent height increases in 2005, 2011, and 2020.
