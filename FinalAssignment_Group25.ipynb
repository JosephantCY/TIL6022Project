## 1. Introduction

In 2003, drivers of specific vehicle types had to pay a few British pounds to enter the area with a vehicle in the centre of London during specific hours. The goal of this charge was to reduce the vehicular activities in the centre of London in order to get rid of the congestion. Therefore, it is interesting to examine if the congestion charge is successful in achieving this goal. This leads to the following research question:

    “How does the congestion charge in London impact the vehicular activities within the congestion charge zone?”

To answer this question, there will be a look into the numbers of different vehicle types and the amount of kilometres travelled by cars within the congestion charge zone. This leads to the following two subquestions:

    1) "To what extent does the congestion charge in London affect the numbers of different vehicle types within the congestion charge zone?"
    2) "To what extent does the congestion charge in London affect the travelled kilometres by cars within the congestion charge zone?"

The research for answering these questions spans from 2000 to 2021, encompassing the period before and after the introduction of the congestion charge in 2003. This approach allows an examination of the vehicular activities preceding the implementation of the charge and the periodically increase of the price of the charge over the years.

## 2. Background research and conceptualisation

The congestion charge in London was established at the beginning of 2003, requiring drivers to pay a specific fee to enter a designated area of London during certain weekday hours. Over the years, the price of the charge has progressively increased (Dunt, 2023), as illustrated in Table 1. The designated zone for this congestion charge encompasses Westminster, Lambeth, Southwark, City of London, Tower Hamlets, Islington and Camden (Tang, 2021).


The vehicles for which the congestion charge in the centre London applies are cars, taxis, Large Goods Vehicles (LGV) and Heavy Goods Vehicles (HGV). The primary objective of this charge is thus to reduce the appeal of these vehicle types, and to increase the appeal of two-wheeled motor vehicles, buses, coaches and pedal cycles (Beevers & Carslaw, 2005). This is deemed essential in addressing traffic congestion, improving air quality, combating climate change, and reducing noise pollution in London (Leape, 2006).

Numerous studies have demonstrated that the implementation of a congestion charge has the potential to curtail both the number of vehicles for which the charge applies and the total distance travelled by vehicles for which the charge applies (Börjesson & Kristoffersson, 2015; Eliasson, 2008; Gibson & Carnovale, 2015; May & Milne, 2000; Seik, 2000). As such, the underlying hypothesis posits that with the introduction of the congestion charge in London in 2003 and the increase of the height of the charge over the years, there will be a gradual reduction over time in the number of cars, taxis, Light Good Vehicles and Heavy Good Vehicles. As a consequence of this, it is expected that the number of two-wheeled motor vehicles, buses, coaches and pedal cycles gradually increases in the congestion zone. Also, it is anticipated that the total travelled distance by car in the congestion zone wil gradually decrease over the years. Finally, it is anticipated that the most substantial changes in both vehicle counts and kilometres travelled will be observed following the charge’s initiation in 2003 and its subsequent height increases in 2005, 2011 and 2020. ### 4.2. Data visualisation and analysis

The graph below illustrates the year-over-year changes in kilometers travelled by cars within the congestion zone from 2000 to 2021. It indicates a gradual decline in the distance travelled by cars over these years except for the decrease in 2020. Consequently, it can be inferred that the congestion charge in London has an impact on the number of kilometers travelled by cars within the congestion zone.

An analysis is conducted to better understand the extent to which the congestion charge impacts the distance travelled by cars within the congestion charge zone. This involves the calculation of the average distance travelled by cars within each congestion pricing period, as well as determining the percentage change in travelled distance compared to the previous period. The results of this analysis are presented in Table 2.

The analysis of Table 2 reveals several key insights:

- During the period 2000-2002 without the congestion charge, there was a decrease of 1.37% in kilometers travelled by cars compared to 2000. Possible reasons for this decline could be attributed to other car-unfriendly measures or public transport stimulance. This percentage likely represents the default decline for the upcoming years if the congestion charge was never introduced.

- In the period 2003-2004, with a £5 congestion charge, the average distance travelled by car is reduced by 5.57% compared to the period without the charge. This percentage is notably higher than the expected default, indicating a significant impact of the congestion charge during this period under the assumption that no other policies or circumstances has an influence.

- Between 2005-2010, as the congestion charge increased by 60% to £8, the average distance travelled by car decreased by 6.20% compared to the £5 charge period. This decrease is significantly higher than the expected default, signifying the congestion charge's influence during this period under the assumption that no other policies or circumstances has an influence.

- With a 25% increase in the congestion charge to £10 during 2011-2013, the average distance travelled by car is reduced by 9.81% compared to the previous period. This reduction is considerably higher than the default, indicating the congestion charge's effect during this period. This conclusion can be made under the assumption that no other policies or circumstances has an influence. Notably, the reduction in kilometers travelled by car was higher during 2011-2013, despite a lower price increase compared to 2005-2010.

- From 2014-2019, with a 15% increase in the congestion charge to £11.50, there was a 2.68% decrease in the average distance travelled by car compared to 2011-2013. This slightly higher percentage signifies a modest impact of the congestion charge during this period under the assumption that no other policies or circumstances has an influence.

- During 2020-2021, with a 30.4% increase in the congestion charge to £15, there was a significant 17.86% decrease in the average distance travelled by car compared to 2014-2019. However, it is crucial to note that his considerable change is likely heavily influenced by the COVID-19 pandemic, and not necessary the congestion charge. For these years it it thus difficult to conclude what the influence is of the congestion charge on the kilometers travelled by car in the congestion zone.
