## 1. Introduction

In 2003, drivers of specific vehicle types had to pay a few British pounds to enter the centre of London during specific hours. The goal of this charge was to reduce the vehicular activities in the centre of London. Therefore, it is interesting to examine if the congestion charge is successful in achieving this goal. This leads to the following research question:

    “How does the congestion charge in London impact the vehicular activities within the congestion charge zone?”

To answer this question, there will be a look into the numbers of different vehicle types and the amount of kilometres travelled by cars within the congestion charge zone. This leads to the following two subquestions:

    1) "To what extent does the congestion charge in London affect the numbers of different vehicle types within the congestion charge zone?"
    2) "To what extent does the congestion charge in London affect the travelled kilometres by cars within the congestion charge zone?"

The research for answering these questions spans from 2000 to 2021, encompassing the period before and after the introduction of the congestion charge in 2003. This approach allows an examination of the vehicular activities preceding the implementation of the charge and the periodically increase of the price of the charge over the years.

## 2. Background research and conceptualisation

The congestion charge in London was established at the beginning of 2003, requiring drivers to pay a specific fee to enter a designated area of London during certain weekday hours. Over the years, the price of the charge has progressively increased (Dunt, 2023), as illustrated in Table 1. The designated zone for this congestion charge encompasses Westminster, Lambeth, Southwark, City of London, Tower Hamlets, Islington and Camden (Tang, 2021).


The vehicles for which the congestion charge in the centre London applies are cars, taxis, Large Goods Vehicles (LGV) and Heavy Goods Vehicles (HGV). The primary objective of this charge is thus to reduce the appeal of these vehicle types, and to increase the appeal of two-wheeled motor vehicles, buses, coaches and pedal cycles (Beevers & Carslaw, 2005). This is deemed essential in addressing traffic congestion, improving air quality, combating climate change, and reducing noise pollution in London (Leape, 2006).

Numerous studies have demonstrated that the implementation of a congestion charge has the potential to curtail both the number of vehicles for which the charge applies and the total distance travelled by vehicles for which the charge applies (Börjesson & Kristoffersson, 2015; Eliasson, 2008; Gibson & Carnovale, 2015; May & Milne, 2000; Seik, 2000). As such, the underlying hypothesis posits that with the introduction of the congestion charge in London in 2003 and the increase of the height of the charge over the years, there will be a gradual reduction over time in the number of cars, taxis, Light Good Vehicles and Heavy Good Vehicles. As a consequence of this, it is expected that the number of two-wheeled motor vehicles, buses, coaches and pedal cycles gradually increases in the congestion zone. Also, it is anticipated that the travelled distance by car in the congestion zone wil gradually decrease over the years. Finally, it is anticipated that the most substantial changes in both vehicle counts and kilometres travelled will be observed following the charge’s initiation in 2003 and its subsequent height increases in 2005, 2011, 2014 and 2020. ### 4.2. Data visualisation and analysis

The graph below illustrates the year-over-year changes in kilometers travelled by cars within the congestion zone from 2000 to 2021. It indicates a gradual decline in the kilometers travelled by car over these years except for the decrease in 2020. Consequently, it can be inferred that in general the congestion charge in London has an impact on the number of kilometers travelled by cars within the congestion zone.

An analysis is conducted to better understand the extent to which the congestion charge impacts the travelled kilometers by cars within the congestion charge zone. This involves the calculation of the percentage change in the travelled kilometres by car compared with the previous year. The results of this analysis are presented in Table 2.

The analysis of Table 2 reveals several key insights:

- During the period 2000-2002 without the congestion charge, there was an average decrease of 1.31% in kilometres travelled by cars. Possible reasons for this decline could be attributed to other car-unfriendly measures and/or public transport measures. This percentage likely represents the default decline for the upcoming years if the congestion charge was never introduced.

- The introduction of the congestion charge in 2003 leads to a reduction of the travelled kilometres by car with 3.40%. This percentage is notably higher than the default decline, indicating a significant impact of the congestion charge under the assumption that no other policies or circumstances has an influence. The effect of the congestion charge can also be observed in 2004 where the decline is 2.04%.

- In 2005, the congestion charge increased by 60% to £8 resulting in a 2.19% decrease in travelled kilometres by cars. This decrease is significantly higher than the expected default, signifying the congestion charge’s influence in this year under the assumption that no other policies or circumstances has an influence. Also, in 2008 and 2010 the decrease is significantly higher than the default decline. This can be the result of the congestion charge but also from other policies.

- With a 25% increase in the congestion charge to £10 in 2011, the travelled kilometres by car is reduced by 2.18% in 2011 and by 4.13% in 2012. This reduction is considerably higher than the default decline, indicating the congestion charge’s effect during these years. This conclusion can be made under the assumption that no other policies or circumstances has an influence.

- In 2014, the congestion charge increased with 15% to £11.50. This resulted in a 0.91% increase in the travelled kilometres by car. Also in 2015, 2016 and 2017 the congestion charge did not result in significant less kilometres travelled by car. Only in 2018, the travelled kilometres by car is reduced by 2.18%, but this is likely not the result of the height increase of the congestion charge. It can thus be concluded that the height increase of the congestion charge in 2014 does not affect the travelled kilometres by car in the congestion zone.

- In 2020, with a 30.4% increase in the congestion charge to £15, there was a significant 20.89% decrease in the travelled kilometres by cars in the congestion zone. However, it is crucial to note that this considerable change is likely heavily influenced by the COVID-19 pandemic, and not necessary the congestion charge. For the years 2020 and 2021 it is thus difficult to conclude what the influence is of the congestion charge on the kilometres travelled by car in the congestion zone.
